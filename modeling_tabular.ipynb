```python
from autogluon.tabular import TabularDataset, TabularPredictor
import pandas as pd
```

```python
df = pd.read_csv("data/datas_cleaned.csv")
df.head()
```

```python
train_data = TabularDataset(df[df["year"] <= 2554].drop(columns=["size_estabish"]))
test_data = TabularDataset(df[df["year"] >= 2556].drop(columns=["size_estabish"]))
```

```python
train_data["value"].describe()
```

```python
predictor = TabularPredictor(label="value", problem_type="regression").fit(
    train_data,
    presets="medium_quality",
)
```

```python
predictor.evaluate(test_data)
```

```python
predictor.leaderboard(test_data)
```

```python
predictor.predict(test_data, model="RandomForestMSE")
```
